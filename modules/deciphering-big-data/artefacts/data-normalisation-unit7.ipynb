```python
import pandas as pd
df = pd.read_excel("DBD_PCOM7E Table.xlsx")
df.head()
```

```python
df.columns.tolist()
```

```python
df.columns = df.columns.str.strip()
```

```python
df[['Student Number', 'Student Name', 'Support', 'Date of Birth']] = (
    df[['Student Number', 'Student Name', 'Support', 'Date of Birth']].ffill()
)
df = df.dropna(subset=['Course Name'])
df.head(10)
```

```python
# Creating unique tables for each entity
students = df[['Student Number', 'Student Name', 'Support', 'Date of Birth']].drop_duplicates()
courses = df[['Course Name', 'Exam Boards']].drop_duplicates()
teachers = df[['Teacher Name']].drop_duplicates()
```

```python
# Creating relationship table between students, courses, and teachers
enrollments = df[['Student Number', 'Course Name', 'Teacher Name', 'Exam Score']].drop_duplicates()
```

```python
students.to_csv("students.csv", index=False)
courses.to_csv("courses.csv", index=False)
teachers.to_csv("teachers.csv", index=False)
enrollments.to_csv("enrollments.csv", index=False)

print("Normalized tables saved successfully!")
```
